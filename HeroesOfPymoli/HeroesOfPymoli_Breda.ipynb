```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
df_purchase_data = pd.read_csv(file_to_load)
df_purchase_data
```

## Player Count

* Display the total number of players


```python
players = df_purchase_data["SN"].unique()
                                 
df_players_count = pd.DataFrame([{"Unique User Count":len(players)}])
df_players_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# # PURCHASING ANALYSIS # #

# Set holders for Purchasing Analysis Summary 
unique_items = len(df_purchase_data["Item Name"].unique())
total_rev = df_purchase_data["Price"].sum()
purchases_count = len(df_purchase_data["Price"])
average_price = total_rev / purchases_count

# Create data frame for Purchasing Analysis Summary
df_purchasing_analysis = pd.DataFrame([{"Number of Unique Items":unique_items, 
                                        "Average Price":average_price,
                                       "Number of Purchases":purchases_count,
                                       "Total Revenue":total_rev}])

# Format Average Price and Total Revenue as currency
df_purchasing_analysis["Average Price"] = df_purchasing_analysis["Average Price"].map("${0:,.2f}".format)
df_purchasing_analysis["Total Revenue"] = df_purchasing_analysis["Total Revenue"].map("${0:,.2f}".format)


df_purchasing_analysis

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# # Gender Demographics Summary # #

# Creates a data frame for storing just purchase data, screen names, and gender
df_purchaser_genders = df_purchase_data[["SN", "Gender", "Price"]].copy()

# Creates a data frame for unique player genders
df_unique_player_genders = df_purchaser_genders.drop_duplicates(subset=["SN", "Gender"], keep="first")
unique_player_count = len(df_unique_player_genders)


# # Stores values to populate Gender Demographics Summary
gender_counts = df_unique_player_genders["Gender"].value_counts()

df_gender_demographics = pd.DataFrame({"Unique Players by Gender": gender_counts, 
                                 "% of Total": gender_counts/unique_player_count})

# Format % of Total as percentage
df_gender_demographics["% of Total"] = df_gender_demographics["% of Total"].map("{0:.1%}".format)



```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
## Purchasing Analysis (Gender) ##

# Create data frame for analyzing purchases by gender
df_grouped_purchaser_genders = df_purchaser_genders.groupby(["Gender"])

# Create holders for key summary items
purchases_count_by_gender = df_grouped_purchaser_genders["Price"].count()
average_purchase_price_by_gender = df_grouped_purchaser_genders["Price"].mean()
purchases_sum_by_gender = df_grouped_purchaser_genders["Price"].sum()

# Create data frame for listing Purchasing Analysis (Gender) Summary
df_purchasing_analysis_gender = pd.DataFrame({"Purchases Count": purchases_count_by_gender, 
                                              "Avg Purchase Price": average_purchase_price_by_gender,
                                             "Total Purchase Value": purchases_sum_by_gender,
                                              "% of Total Revenue": purchases_sum_by_gender/total_rev,
                                             "Avg Purchases / Person":purchases_count_by_gender/gender_counts})

# Sets formatting
df_purchasing_analysis_gender["Avg Purchase Price"] = df_purchasing_analysis_gender["Avg Purchase Price"].map("${0:,.2f}".format)
df_purchasing_analysis_gender["Total Purchase Value"] = df_purchasing_analysis_gender["Total Purchase Value"].map("${0:,.2f}".format)
df_purchasing_analysis_gender["% of Total Revenue"] = df_purchasing_analysis_gender["% of Total Revenue"].map("{0:.1%}".format)
df_purchasing_analysis_gender["Avg Purchases / Person"] = df_purchasing_analysis_gender["Avg Purchases / Person"].map("{0:,.1f}".format)

df_purchasing_analysis_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
## Age Demographics Summary ##

# Create a data frame for storing just purchase data, screen names, and age
df_purchaser_ages = df_purchase_data[["SN", "Age", "Price"]].copy()

# Create a data frame for unique player genders
df_unique_player_ages = df_purchaser_ages.drop_duplicates(subset=["SN", "Age"], keep="first")

# Create bins for storing unique player ages, and pushes those bins to a new list
age_bins = (0,10,15,20,25,30,35,40,100)
age_labels = ("<10","10-14","15-19","20-24","25-29","30-34","35-39","40+")
age_list_binned = pd.cut(df_unique_player_ages["Age"], bins=age_bins, right=False, labels=age_labels)

# Create new data frame for summarizing age demographics
df_unique_age_counts = pd.DataFrame({"Unique User Count":df_unique_player_ages["SN"],
                                    "Age Bin":age_list_binned})

df_grouped_unique_age_counts = df_unique_age_counts.groupby(["Age Bin"]).count()
df_grouped_unique_age_counts["% of Total"] = df_grouped_unique_age_counts["Unique User Count"] / unique_player_count

# Format % of total
df_grouped_unique_age_counts["% of Total"] = df_grouped_unique_age_counts["% of Total"].map("{0:.2%}".format)
df_grouped_unique_age_counts
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
## Purchasing Analysis (Age) ##

# Pull in all purchase data and add a column for age bin

all_purchasers_age_bins = pd.cut(df_purchase_data["Age"], bins=age_bins, 
                                 right=False, labels=age_labels)

df_purchaser_ages["Age Bin"] = all_purchasers_age_bins

df_grouped_purchaser_age_counts = df_purchaser_ages.groupby(["Age Bin"]).count()
df_grouped_purchaser_age_sums = df_purchaser_ages.groupby(["Age Bin"]).sum()
df_grouped_purchaser_age_means = df_purchaser_ages.groupby(["Age Bin"]).mean()

# Create holders for key summary items
purchase_count_by_age = df_grouped_purchaser_age_counts["SN"]
purchase_sum_by_age = df_grouped_purchaser_age_sums["Price"]
purchase_avg_by_age = df_grouped_purchaser_age_means["Price"]
avg_purchase_by_person = purchase_sum_by_age / df_grouped_unique_age_counts["Unique User Count"]

# Create data frame for Purchasing Analysis (Age)
df_purchasing_analysis_age = pd.DataFrame({"Purchase Count":purchase_count_by_age,
                                          "Average Purchase Price": purchase_avg_by_age,
                                          "Total Purchase Value": purchase_sum_by_age,
                                           "% of Total Revenue": purchase_sum_by_age / total_rev,
                                          "Avg Total Purchase / Person": avg_purchase_by_person})

# Apply formatting
df_purchasing_analysis_age["Average Purchase Price"] = df_purchasing_analysis_age["Average Purchase Price"].map("${0:,.2f}".format)
df_purchasing_analysis_age["Total Purchase Value"] = df_purchasing_analysis_age["Total Purchase Value"].map("${0:,.2f}".format)
df_purchasing_analysis_age["% of Total Revenue"] = df_purchasing_analysis_age["% of Total Revenue"].map("{0:.1%}".format)
df_purchasing_analysis_age["Avg Total Purchase / Person"] = df_purchasing_analysis_age["Avg Total Purchase / Person"].map("${0:,.2f}".format)


df_purchasing_analysis_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
## Top Spenders Analysis ##

# Establishes holders for key summary items

df_purchase_total_sums_by_SN = df_purchase_data.groupby(["SN"]).sum()
sorted_df_purchase_total_sums_by_SN = df_purchase_total_sums_by_SN.sort_values("Price", ascending=False)

purchase_count_by_sn = df_purchase_data["SN"].value_counts()
purchase_sum_by_sn = sorted_df_purchase_total_sums_by_SN["Price"]

# Create data frame for top spender summary
df_top_spenders = pd.DataFrame({"Purchase Count": purchase_count_by_sn,
                               "Avg Purchase Price": purchase_sum_by_sn / purchase_count_by_sn,
                               "Total Purchase Value": purchase_sum_by_sn})

# Sort top spender summary by purchase value
df_top_spenders = df_top_spenders.sort_values("Total Purchase Value",ascending=False)

# Apply formatting
df_top_spenders["Avg Purchase Price"] = df_top_spenders["Avg Purchase Price"].map("${0:,.2f}".format)
df_top_spenders["Total Purchase Value"] = df_top_spenders["Total Purchase Value"].map("${0:,.2f}".format)


df_top_spenders.head(10)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
## Most Popular Items ##

# Establish holders for key summary items

items = df_purchase_data[["Item ID", "Item Name", "Price"]]
grouped_items = items.groupby(["Item ID", "Item Name"])

purchase_count_by_item = grouped_items["Item ID"].count()
total_purchase_value_by_item = grouped_items["Price"].sum()
item_price = total_purchase_value_by_item / purchase_count_by_item

# Create data frame for Most Popular Items Summary
df_most_popular_items = pd.DataFrame({"Purchase Count": purchase_count_by_item,
                                     "Item Price": item_price,
                                     "Total Purchase Value": total_purchase_value_by_item})

df_most_popular_items = df_most_popular_items.sort_values("Purchase Count", ascending=False)

# Apply formatting
df_most_popular_items["Item Price"] = df_most_popular_items["Item Price"].map("${0:.2f}".format)
df_most_popular_items["Total Purchase Value"] = df_most_popular_items["Total Purchase Value"].map("${0:.2f}".format)


df_most_popular_items.head(10)

```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
## Most Profitable Items ##

# Creates a copy of the most popular items table, then sorts by total purchase value

df_most_profitable_items = pd.DataFrame({"Purchase Count": purchase_count_by_item,
                                     "Item Price": item_price,
                                     "Total Purchase Value": total_purchase_value_by_item})

df_most_profitable_items = df_most_profitable_items.sort_values("Total Purchase Value", ascending=False)

# Apply formatting
df_most_profitable_items["Item Price"] = df_most_profitable_items["Item Price"].map("${0:.2f}".format)
df_most_profitable_items["Total Purchase Value"] = df_most_profitable_items["Total Purchase Value"].map("${0:.2f}".format)

df_most_profitable_items
```
